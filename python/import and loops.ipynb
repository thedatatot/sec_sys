```python
# Property Check Functions
# starts with is******
#  'isalnum',
#  'isalpha',
#  'isascii',
#  'isdecimal',
#  'isdigit',
#  'isidentifier',
#  'islower',
#  'isnumeric',
#  'isprintable',
#  'isspace',
#  'istitle',
#  'isupper',
    
s = 'rahul'
print(s.isalnum())
print(s.isalpha())
print(s.isascii())
print(s.isdecimal())
print(s.isdigit())
print(s.isidentifier())
print(s.islower())
print(s.isnumeric())
print(s.isprintable())
print(s.isspace())
print(s.istitle())
print(s.isupper())
```

```python
#  'capitalize',  'casefold',  'center',  'count',
#  'encode',  'endswith',  'expandtabs',  'find',
#  'format',  'format_map',  'index',  'join',
#  'ljust',  'lower',  'lstrip',  'maketrans',
#  'partition',  'replace',  'rfind',  'rindex',
#  'rjust',  'rpartition',  'rsplit',  'rstrip',
#  'split',  'splitlines',  'startswith',  'strip',
#  'swapcase',  'title',  'translate',  'upper',  'zfill'

s = 'rahul kumar'
print(s.capitalize())
print(s.casefold())
print(s.center(24,'*'))
print(s.count('rahul'), s.encode(), s.endswith('h'), s.expandtabs(), sep = ' $ ' )
print(s.swapcase(), s.upper())
```
