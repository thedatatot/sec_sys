```python
# WAP to input a string and print ascii value of each character

# c = 'a'
# print(ord(c))
# print(c[0])

c = input('Enter any string : ')
i = 0
while i < len(c):
    print(ord(c[i]))
    i += 1
```

```python
# WAP to input a string and find sum of ascii value of all characters

c = input('Enter any string : ')

i = 0
s = 0

while i < len(c):
    n = ord(c[i])
    s = s + n
    i += 1

    
print(s)
```
